## Initializing BigQuery client

```python
from google.cloud import bigquery

project_id = "icoads-time-series-analysis"
client = bigquery.Client(project=project_id)
```

## Data Copy from Open dataset to project workspace

```python
query = """
CREATE OR REPLACE TABLE `icoads-time-series-analysis.icoads_data.icoads_raw_aggregated` AS
SELECT * FROM `bigquery-public-data.noaa_icoads.icoads_core_2001_2004`
UNION ALL
SELECT *
FROM `bigquery-public-data.noaa_icoads.icoads_core_*`
WHERE _TABLE_SUFFIX BETWEEN '2005' AND '2017'
"""

query_job = client.query(query)
query_job.result()
print("Table created successfully!")
```

## Testing table availability

```python
dataset_ref = client.dataset(dataset_id='icoads_data')
tables = list(client.list_tables(dataset_ref))
for table in tables:
  print(table.table_id)
```

```python
df = client.query('''
SELECT sea_surface_temp FROM `icoads-time-series-analysis.icoads_data.icoads_raw_aggregated`
WHERE year = 2005
''').to_dataframe()

display(df)
```
